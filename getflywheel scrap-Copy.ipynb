```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
from time import sleep
import csv
import xlsxwriter
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as ec
import pandas as pd
```

```python
#directory-item__card-agency
driver= webdriver.Chrome()
#sleep(2)
print('01010101010101')
url = 'https://getflywheel.com/agency-partners-directory/'
driver.get(url)
print('11111111111')
page_source = BeautifulSoup(driver.page_source)
    #print('11111111111')
profiles = page_source.find_all("div",{"class":"directory-item__card-agency"})
#profiles[0]
all_profile_URL = []
for profile in profiles:
        a = profile.select("a")
        for i in a :
            profile_URL = i.get('href')
            if profile_URL not in all_profile_URL:
                all_profile_URL.append(profile_URL)
                print(profile_URL)

all_profile_URL
print(len(all_profile_URL))

#https://getflywheel.com/agency-partners-directory/?sf_paged=


num_of_pages = 70
#url_all_page = all_profile_URL
#all_profile_URL = []

for i  in range(num_of_pages):
    url_all_page = []
    i+=2
    x = str(i)
    driver1= webdriver.Chrome()
    #sleep(2)
    url2 = 'https://getflywheel.com/agency-partners-directory/?sf_paged='+ x +''
    #sleep(5)
    driver1.get(url2)
    #sleep(5)
    page_source1 = BeautifulSoup(driver1.page_source)
    print('--------')
    #sleep(5)
    profiles = page_source1.find_all("div",{"class":"directory-item__card-agency"})
    #sleep(5)
    print('----1----')
    for profile in profiles:
        #print('---2-----')
        a = profile.select("a")
        #print('-----3---')
        for i in a :
            profile_URL = i.get('href')
            if profile_URL not in url_all_page:
                url_all_page.append(profile_URL)
                print(profile_URL)
        
    all_profile_URL = all_profile_URL + url_all_page
        
    sleep(2)
    print(len(all_profile_URL))   
    
print(len(all_profile_URL))    
output = pd.DataFrame()    
output = output.append(all_profile_URL , ignore_index=True)
output.to_excel("xxxxxxx.xlsx")
print('\n-----------Succesfully get all url pages------\n')
```
